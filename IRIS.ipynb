```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris
```

# 1.WHAT IS THE DISTRIBUTION OF PETAL LENGTHS IN THE DATASET?

```python
sns.histplot(x='petal_length',data=iris,kde=True,color='aqua')
plt.show()
```

# 2.HOW DOES PETAL LENGTH VARY ACROSS DIFFERENT SPECIES?

```python
sns.violinplot(x='species',y='petal_length',data=iris,color='teal')
plt.show()
```

# 3.WHAT IS THE RELATIONSHIP BETWEEN PETAL LENGTH AND PETAL WIDTH, COLORED BY SPECIES?

```python
sns.scatterplot(x='petal_length',y='petal_width',data=iris,hue='species')
plt.show()
```

# 4.HOW ARE ALL NUMERICAL FEATURES RELATED TO EACH OTHER?

```python
corr=iris.drop('species',axis=1).corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

# 5.HOW DOES SEPAL WIDTH VARY ACROSS THE THREE SPECIES?

```python
sns.boxplot(x='species',y='sepal_width',data=iris,color='turquoise')
plt.show()
```

# 6.WHAT’S THE RELATIONSHIP BETWEEN SEPAL LENGTH AND PETAL LENGTH, COLORED BY SPECIES?

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
plt.show()
```

# 7.WHAT IS THE AVERAGE PETAL WIDTH FOR EACH SPECIES?

```python
sns.barplot(x='species',y='petal_width',data=iris,color='turquoise')
plt.show()
```

# 8. HOW DOES SEPAL WIDTH RELATE TO SEPAL LENGTH, COLORED BY SPECIES?

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,hue='species')
plt.show()
```

# 9.WHAT IS THE DISTRIBUTION OF SEPAL WIDTH ACROSS ALL SPECIES COMBINED?

```python
sns.histplot(x='sepal_width',kde=True,color='teal',data=iris)
plt.show()
```

```python
sns.pairplot(data=iris, hue='species', palette='husl')
plt.show()
```

# 🌸 Exploratory Data Analysis (EDA) on the Iris Dataset using Seaborn

## 📊 Dataset Overview:
- **Total Samples**: 150
- **Target**: `species` (Setosa, Versicolor, Virginica)
- **Features**:
  - Sepal Length (cm)
  - Sepal Width (cm)
  - Petal Length (cm)
  - Petal Width (cm)

---

## 🔍 Summary of 10 EDA Queries:

1. **Distribution of Petal Length**  
   → `histplot` + KDE showed bimodal distribution (Setosa vs. others)

2. **Petal Length by Species**  
   → `boxplot` revealed median differences and IQR per class

3. **Petal Length vs Petal Width**  
   → `scatterplot` showed clear species separation (strong correlation)

4. **Correlation Matrix**  
   → `heatmap` showed strong correlation between petal length & petal width

5. **Sepal Width by Species**  
   → `violinplot` revealed distribution shape and density per species

6. **Sepal Length vs Petal Length**  
   → `scatterplot` showed linear trends and group clustering

7. **Average Petal Width per Species**  
   → `barplot` showed virginica has the highest mean width

8. **Sepal Width vs Sepal Length**  
   → `scatterplot` showed Setosa clustering away from others

9. **Distribution of Sepal Width**  
   → `histplot` with KDE highlighted slight left-skewed shape

10. **Pairwise Plot of All Features**  
    → `pairplot` visualized all numerical feature combinations by species

---

## 🧠 Insights:
- Petal features (length & width) offer **strong class separability**
- Sepal features have **weaker separation**
- Setosa is consistently distinct, while Versicolor and Virginica overlap more
- Petal length is **strongly correlated** with petal width
